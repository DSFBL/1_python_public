```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
df = pd.read_csv('populations.txt', sep='\t')
df.head()
```

```python
df
```

```python
plt.plot(df['year'], df['lynx']); #lineplot
```

```python
plt.bar(df['year'], df['lynx']); #barplot
```

```python
fig, ax = plt.subplots(1, 1, figsize=(7.5, 5))

ax.plot(df.year, df.lynx, color='lime', linestyle='--');

ax.locator_params(integer=True) # сделали года целыми числами

ax.set_xlabel('Year', fontsize=12)
ax.set_ylabel('Lynx', fontsize=12)
ax.set_title('Lynx by year', fontsize=15)

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
```

```python
fig, ax = plt.subplots(2, 2, figsize=(15, 10))

ax[0][0].plot(df.year, df.lynx, color='lime', linestyle='--')

ax[0][0].locator_params(integer=True) # сделали года целыми числами

ax[0][0].set_xlabel('Year', fontsize=12)
ax[0][0].set_ylabel('Lynx', fontsize=12)
ax[0][0].set_title('Lynx by year', fontsize=15)

ax[0][1].plot(df.year, df.hare, color='indigo', linestyle='--')
ax[0][1].set_xlabel('Year', fontsize=12)
ax[0][1].set_ylabel('Hare', fontsize=12)
ax[0][1].set_title('Hare by year', fontsize=15)
ax[0][1].locator_params(integer=True)

ax[1][0].plot(df.year, df.carrot, color='maroon', linestyle='--')
ax[1][0].set_xlabel('Year', fontsize=12)
ax[1][0].set_ylabel('Carrot', fontsize=12)
ax[1][0].set_title('Carrot by year', fontsize=15)
ax[1][0].locator_params(integer=True)

ax[1][1].plot(df.year, df.lynx, color='lime', linestyle='--', label='lynx')
ax[1][1].plot(df.year, df.hare, color='indigo', linestyle='--', label='hare')
ax[1][1].plot(df.year, df.carrot, color='maroon', linestyle='--', label='carrot')
ax[1][1].legend(loc=1, fontsize=10, frameon=False);
```

```python
df_crime = pd.read_csv('crimeRatesByState2005.tsv', sep='\t')
df_crime.head()
```

```python
fig, ax = plt.subplots(figsize=(15, 10))

crime_scatter = ax.scatter(df_crime['murder'], df_crime['burglary'], s=df_crime['population']/40000,
                           alpha=0.5, c=df_crime['Robbery'], cmap='cool', linewidth=2)

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

for i, state in enumerate(df_crime['state']):
    ax.annotate(state, (df_crime['murder'][i], df_crime['burglary'][i]), fontsize=8)

plt.colorbar(crime_scatter, label='Robbery')
ax.set_xlabel('murder')
ax.set_ylabel('burglary');
```

```python
fig.savefig('crimes.pdf')
```
