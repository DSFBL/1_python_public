```python
def get_books_info(num_page):
    url = f'https://books.toscrape.com/catalogue/page-{num_page}.html'
    response = requests.get(url)
    tree = BeautifulSoup(response.content, 'html.parser')
    all_books = tree.find_all('article', {'class': 'product_pod'})

    books_info = []
    for book in all_books:
        d = {'title': book.h3.a.get('title'),
            'stars_rating': book.p.get('class')[1],
            'price': book.find('p', {'class': 'price_color'}).text,
            'url': book.a.get('href')}
        books_info.append(d)
    return books_info
```

```python
num_page = 76
f'https://books.toscrape.com/catalogue/page-{num_page}.html'
```

```python
books_data = []
for i in range(1, 51):
    books_data += get_books_info(i)
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(books_data)
```

```python
df
```

```python
def get_rating(string):
    if string == 'One':
        return 1
    elif string == "Two":
        return 2
    elif string == "Three":
        return 3
    elif string == "Four":
        return 4
    elif string == "Five":
        return 5
    else:
        return -999
```

```python
df['stars_rating'] = df['stars_rating'].apply(get_rating)
```

```python
df['stars_rating'].hist();
```

```python
def get_price(string):
    return float(string[1:])
```

```python
df['price_pounds'] = df['price'].apply(get_price)
```

```python
import numpy as np
```

```python
df['price_rubles'] = np.round(df['price_pounds'] * 102.89, 2)
```

```python
df.drop(['price'], axis=1, inplace=True)
```
