### 1. Знакомство со словарями

Изучим такой `тип данных`, как `словари(dict)`.

Несмотря на простое название, работа со словарями обычно сложна для понимания студентов.
Однако если вы уже так много изучили и дошли до темы словаря, то вы справитесь и с ней! И мы поможем вам в этом :)

`Словарь` – неупорядоченная коллекция произвольных объектов с доступом по ключу, заключенная в фигурные скобки.

Ниже будет более подробно объяснено данное определение `словаря`.

Словарь состоит из объектов, у каждого из которых есть ключ, предоставляющий доступ к самому объекту. Пока что сложно?:) Скоро станет понятно.

Словарь - это множество пар такого вида: `'КЛЮЧ': Объект`.

В отличии от строк, в словарях объекты `не упорядочены`. То есть в каком порядке размещены данные пары `'КЛЮЧ': Объект` внутри словаря не имеет значения.

Представим, что в продуктовом магазине продаются: арбузы, апельсины, лимоны, дыни и виноград. У каждого товара есть своя цена, которая изображена на ценнике. Для того чтобы узнать цену, нам нужно обратиться к определенному товару.
Например: перед тем, как купить лимон, покупатель берет его в руки, оценивает и смотрит на ценник.

В Python имена товаров (`'арбуз'`, `'апельсин'`, `'лимон'`, `'дыня'`, `'виноград'`) будут называться `ключами`.

Тогда цены товаров (`70`, `50`, `60`, `90`, `80`) будут являться `объектами словаря`.

Чтобы узнать цену товара, нам нужно знать имя товара. То есть чтобы узнать определенный `объект словаря`, мы должны знать его `ключ`.

Давайте попробуем написать такой словарь из товаров и их цен для Python:

```python
shop = {'арбуз': 70 , 'апельсин': 50 , 'лимон': 60 , 'дыня': 90 , 'виноград': 80}
```

- Как уже было написано выше,  коллекция объектов и ключей к ним записывается в `фигурных скобках`: { ... }.

- Между каждым объектом (70, 50, 60, 90, 80) и ключу к нему ('арбуз', 'апельсин', 'лимон', 'дыня', 'виноград') ставится `знак двоеточия`.

- Словарь представляет собой коллекцию пар `'ключ': объект`, и все эти пары разделены `запятой`.

```python
print(type(shop))
```

### 2. Операции со словарями

Представим, что мы хотим узнать цену арбуза:

```python
arbuz = shop['арбуз']
print(arbuz)
```

Работает! Также мы можем узнать цены других товаров:

```python
print('Цена апельсина =', shop['апельсин'])
print('Цена лимона =', shop['лимон'])
```

Давайте проверим, что будет, если мы попробуем узнать цену несуществующего товара:

```python
print('Цена яблока =', shop['яблоко'])
```

Очевидно, что Python выдает ошибку и пишет: 'яблоко'; ссылаясь на то, что такого ключа нет в словаре.

Мы можем заранее проверить, существует ли такой ключ в словаре:

```python
'яблоко' in shop
```

```python
'апельсин' in shop
```

Однако мы можем добавить новый объект и ключ к нему в словарь.

```python
shop['яблоко'] = 50
print(shop)
```

Также с помощью операции `del` мы можем удалить объект и ключ к нему из словаря. Пусть в нашем магазине выкупили все яблоки:

```python
del shop['яблоко']
print(shop)
```

В данном примере с продуктовым магазином ключи словаря были представлены в виде `строк` ('арбуз', 'апельсин', 'лимон', 'дыня', 'виноград'), а объекты - в виде `чисел`  (70, 50, 60, 90, 80). 

Однако это не всегда так. Давайте рассмотрим различные примеры таких альтернативных словарей:

```python
myDict = {1: 1, 5: 25, 'three': 9,2: 4}
print(myDict)
```

```python
myDict = {'Россия': 'Москва', 'Казахстан': 'Астана', 'Италия': 'Рим', 'Бразилия': 'Бразилиа', 'Африка': None}
print(myDict)
```

Также в словаре объектом может служить другой словарь. Например, создадим словарь из континентов, стран на этих континентах и столиц, и назовем этот словарь `world`:

```python
world = {
    'Евразия': {
        'Россия': 'Москва',
        'Казахстан': 'Астана',
        'Италия': 'Рим',
        'Индия': 'Нью-Дели' 
    },
    'Африка': {
        'Алжир': 'Алжир',
        'Египет': 'Каир'
    },
    'Антарктида': None
}
```

```python
print(world)
```

```python
print(world['Евразия'])
```

```python
print(world['Африка'])
```

```python
print(world['Антарктида'])
```

Также мы можем отсортировать словарь по алфавиту с помощью операции `sorted`:

```python
print(sorted(world))
```

```python
sorted(world['Евразия'])
```

### 3. Методы словарей

Теперь изучим методы - функции, которые применяютя к объекту в данном случае к словарю. Методы словарей облегчают работу с ними, они выполняют распространненые операции над словарями и делают программный код значительно короче.

Метод `copy` возвращает копию существующего словаря. Например, создадим копию словаря `shop`:

```python
shop2 = shop.copy()
print(shop2)
```

Метод `clear` очищает словарь:

```python
shop2.clear()
print(shop2)
```

Метод `keys` возвращает всю ключи, хранящиеся в словаре, то есть названия товаров в магазине:

```python
shop.keys()
```

Метод `values` возращает все объекты словаря, то есть цены товаровв магазине:

```python
shop.values()
```
